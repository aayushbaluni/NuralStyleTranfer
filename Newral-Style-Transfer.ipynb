```python
import tensorflow_hub as hub
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
```

```python
import numpy as np
```

```python
model=hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
def load_image(path):
    img=tf.io.read_file(path)
    img=tf.image.decode_image(img,channels=3)
    img=tf.image.convert_image_dtype(img,tf.float32)
    img=img[tf.newaxis,:]
    return img
```

```python
style_img=load_image('image1.jpg')
```

```python
content_img=load_image('image2.jpg')
```

```python
plt.imshow(np.squeeze(style_img))
plt.show()
```

```python
plt.imshow(np.squeeze(content_img))
plt.show()
```

```python
styled_img=model(tf.constant(content_img),tf.constant(style_img))[0]
```

```python
plt.imshow(np.squeeze(styled_img))
plt.show()
```
